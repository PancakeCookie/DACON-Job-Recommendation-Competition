## 이상치 처리

```python
for col in know_2018:
        know_2018[col].replace(' ', 0, inplace = True)
```

```python
str_li =['bq4_1a','bq4_1b','bq4_1c','bq5_2', 'bq21','bq28_1','bq29','bq30','bq31', 'bq32','bq33','bq37_1']
```

```python
know_2018 = know_2018.fillna(0)
```

```python
know_2018.isnull().sum()[100:]
```

```python
for i in know_2018.columns : 
    if i not in str_li : 
        know_2018[i] = know_2018[i].astype(int)
    else :
        know_2018[i]=know_2018[i].astype(str)
```
